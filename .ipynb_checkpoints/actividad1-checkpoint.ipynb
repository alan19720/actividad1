```python
# Espacio para realizar la importación de librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ubicacionArchivo = "Calificaciones_Cafe.csv"
# Carga de un CSV a un archivo tipo DataFrame por medio de pandas.
df = pd.read_csv(ubicacionArchivo)
```

```python
#vizualizamos los datos
display(df)
```

```python
print("\nInformación general del DataFrame: \n\n")
print(df.info())

# Contabiliza el número de columnas numéricas
num_columnas_numericas = len(df.select_dtypes(include=['number']).columns)

# Contabiliza el número de columnas de tipo objeto
num_columnas_objeto = len(df.select_dtypes(include=['object']).columns)

# Contabiliza la cantidad de valores nulos por columna
nulos_por_columna = df.isnull().sum().sort_values(ascending=False)

# Muestra los resultados

print("\nNúmero de columnas numéricas:", num_columnas_numericas)
print("\nNúmero de columnas de tipo objeto:", num_columnas_objeto)
print("\n\nCantidad de valores nulos por columna:\n")
print(nulos_por_columna)
```
