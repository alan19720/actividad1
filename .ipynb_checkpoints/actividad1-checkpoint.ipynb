```python
# Espacio para realizar la importación de librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ubicacionArchivo = "Calificaciones_Cafe.csv"
# Carga de un CSV a un archivo tipo DataFrame por medio de pandas.
df = pd.read_csv(ubicacionArchivo)
```

```python
#vizualizamos los datos
display(df)
```

```python
print("\nInformación general del DataFrame: \n\n")
print(df.info())

# Contabilizamos el número de columnas numéricas
num_columnas_numericas = len(df.select_dtypes(include=['number']).columns)

# Contabilizamos el número de columnas de tipo objeto
num_columnas_objeto = len(df.select_dtypes(include=['object']).columns)

# Contabilizamos la cantidad de valores nulos por columna
nulos_por_columna = df.isnull().sum().sort_values(ascending=False)

# Muestramos los resultados

print("\nNúmero de columnas numéricas:", num_columnas_numericas)
print("\nNúmero de columnas de tipo objeto:", num_columnas_objeto)
print("\n\nCantidad de valores nulos por columna:\n")
print(nulos_por_columna)
```

```python
#PASO 2
# encontramos el valor máximo en la columna "total_cup_points"
max_puntaje = df['total_cup_points'].max()

# Imprimimos el resultado
print("El mayor puntaje en 'total_cup_points' es:", max_puntaje)
```

```python
#utilizamos
# Utilizamos el método value_counts() para contar la frecuencia de cada categoría en la columna "species"
histograma_pandas = df['species'].value_counts()

# Creamos una grafica para el histograma utilizando Pandas
histograma_pandas.plot(kind='bar')
plt.title('Histograma de Species (Pandas)')
plt.xlabel('Species')
plt.ylabel('Frecuencia')

# vizualizamos la gráfica
plt.show()
```

```python
# Creamos un histograma utilizando Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='species', bins=10, kde=True)
plt.title('Histograma de Species (Seaborn)')
plt.xlabel('Species')
plt.ylabel('Frecuencia')

# visualizamos la grafica
plt.show()
```

```python
#paso 4
# Histograma con 10 bins
plt.figure(figsize=(10, 6))
plt.hist(df['total_cup_points'], bins=10, color='skyblue', alpha=0.7)
plt.title('Histograma con 10 Bins')
plt.xlabel('Total Cup Points')
plt.ylabel('Frecuencia')
plt.show()

# Histograma con 50 bins
plt.figure(figsize=(10, 6))
plt.hist(df['total_cup_points'], bins=50, color='salmon', alpha=0.7)
plt.title('Histograma con 50 Bins')
plt.xlabel('Total Cup Points')
plt.ylabel('Frecuencia')
plt.show()

# Histograma con 100 bins
plt.figure(figsize=(10, 6))
plt.hist(df['total_cup_points'], bins=100, color='lightgreen', alpha=0.7)
plt.title('Histograma con 100 Bins')
plt.xlabel('Total Cup Points')
plt.ylabel('Frecuencia')
plt.show()

# Histograma con bins igual al número de valores únicos
unico_bins = len(df['total_cup_points'].unique())
plt.figure(figsize=(10, 6))
plt.hist(df['total_cup_points'], bins=unico_bins, color='lightcoral', alpha=0.7)
plt.title(f'Histograma con {unico_bins} Bins (Valores Únicos)')
plt.xlabel('Total Cup Points')
plt.ylabel('Frecuencia')
plt.show()
```

#paso 5
Analizando las graficas anteriores nos percatamos que la elección del número ideal de bins depende del propósito del análisis y de la naturaleza de los datos. Un número demasiado bajo de bins puede ocultar detalles importantes, mientras que un número demasiado alto puede dificultar el entendimiento y analisis devido a la cantidad de bins.
Tomando en cuenta los factores una cantidad de bins ideal seria entre 10 y 50 bins, pero la forma ideal seria utilizar metodos estadisticos que consideran el tamaño de la muestr4a y la variedad de los datos.

```python
#paso 6
# Creamos un histograma con KDE
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='total_cup_points', bins=30, kde=True, color='skyblue')
plt.title('Histograma con KDE de Total Cup Points')
plt.xlabel('Total Cup Points')
plt.ylabel('Frecuencia')

# Muestramos el gráfico
plt.show()
```

Al analizar y observar el histograma con KDE nos percatamos que se asemeja a una campana y la curva KDE se asemeja bien a esta forma se determina que es una distribucion gaussiana.
